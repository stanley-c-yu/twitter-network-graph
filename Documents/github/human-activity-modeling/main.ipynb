The first 12.8 second window of the super window takes up the first 1024 rows.  Dividing the total numbers of rows by 1024 gives us 155.703125 12.8 sub-windows.  Obviously this number is not ideal, but for now we can opt for a lazy fix by just throwing out the last window.  So instead, we'll just divide the Super Sensor data frame into 115 windows.  For now, sense we're still protoyping, we'll just settle for one.  

Now according to the MUSS Study, for each 12.8 second window, the researchers threw out first and last 0.4 seconds to make room for padding, in case the user is in the middle of a transition from one activity to another. 

So now we need to pare down the first window into a tighter bound by lopping off the first and last 0.4 seconds.

Compute the mean acceleration values over the samples in a 2-second sub-window as the approximated orientation vector. 

This means computing the average of X, the average of Y, and the average of Z.  

```python
# Isolate X, Y, and Z 
sw1X = subwindow1['X_ACCELERATION_METERS_PER_SECOND_SQUARED']
sw1Y = subwindow1['Y_ACCELERATION_METERS_PER_SECOND_SQUARED']
sw1Z = subwindow1['Z_ACCELERATION_METERS_PER_SECOND_SQUARED']
# Calculate Average of sw1 X, Y, and Z 
mean_accel_x = np.mean(np.asarray(sw1X))
mean_accel_y = np.mean(np.asarray(sw1Y))
mean_accel_z = np.mean(np.asarray(sw1Z))
print("Mean Acceleration Value of X: ", round(mean_accel_x, 4), " | Mean Acceleration Value of Y: ", round(mean_accel_y, 4), 
      " | Mean Acceleration Value of Z: ", round(mean_accel_z, 4))
approx_orient_vec = [mean_accel_x, mean_accel_y, mean_accel_z]
approx_orient_vec = np.asarray(approx_orient_vec)
approx_orient_vec = np.reshape(approx_orient_vec, (3,1))
print(approx_orient_vec)
print("Dimensions: ", np.shape(approx_orient_vec))
```

Now we need to normalize the mean acceleration values to have unit gravity

```python
import math 
# Calculate denominator 
denominator = math.sqrt(mean_accel_x**2 + mean_accel_y**2 + mean_accel_z**2)
# Normalize Approximated Orientation Vector 
normalized_aov = approx_orient_vec/denominator 
print(normalized_aov)
print("Length of vector: ", np.linalg.norm(normalized_aov))
```

Compute the angles between the X, Y, and Z axis directions and the direction of the gravity as the orientation angles (from 0 10 180): 

```python
orient_angles = np.arccos(normalized_aov)
print(orient_angles)
```

Next step: replicate process for all six sub-windows.  

```python
def filterSensor(sensor_data, ground_truth_data):
    lower_bound = ground_truth_data["START_TIME"].min()
    upper_bound = ground_truth_data["STOP_TIME"].max() 
    query = "SELECT * FROM sensor_one WHERE HEADER_TIME_STAMP BETWEEN '" + lower_bound + "' AND '" + upper_bound + "'"
 
    filteredSensor = pysqldf(query)
    return filteredSensor 

filteredData = filterSensor(sensor_one, ground_truths)
filteredData.head()
```

```python
from datetime import datetime  
from datetime import timedelta  

lower_bound = ground_truths["START_TIME"].min()
print(type(pd.to_datetime(lower_bound)) )
print(pd.to_datetime(lower_bound))
pd.to_datetime(lower_bound) + timedelta(seconds=12.8)
# starting from min time stamp in filtered data
# while below upper threshold, as discussed earlier
#     add 12.8 seconds and append to empty list
#     repeat until threshold is reached 
# for eac pair of values at index i and j
#     subset the dataframe using those indicies 
#     append the subset to something like an empty list 
#     increment i and j in such a way that j = i + 1 
#     repeat until end of list is reached.  
# return sub-windows 
```

```python
filteredData = filterSensor(sensor_one, ground_truths)

time_stamps = []
lower_bound = pd.to_datetime(ground_truths["START_TIME"].min())
print("Lower bound as defined by ground truths: ", lower_bound)
upper_bound = pd.to_datetime(ground_truths["STOP_TIME"].max())
print("Upper bound as defined by ground truths: ", upper_bound)
#filteredData = pd.to_datetime(filteredData["HEADER_TIME_STAMP"])

marker = lower_bound
time_stamps.append(marker)
while marker < upper_bound: 
    marker = marker + timedelta(seconds=12.8)
    time_stamps.append(marker)
print(len(time_stamps))
doce_bounds = []
for i in range(len(time_stamps)): 
    bound1 = time_stamps[i]
    doce_bounds.append(bound1)
    j = i + 1 
    if j == len(time_stamps): 
        break 
    else:
        bound2 = time_stamps[j]
        doce_bounds.append(bound2)
```

```python
filteredData["HEADER_TIME_STAMP"] = pd.to_datetime(filteredData["HEADER_TIME_STAMP"])
bound1 = str(doce_bounds[0])
print(bound1)
bound2 = str(doce_bounds[2])
print(bound2)
filteredData[(filteredData["HEADER_TIME_STAMP"] >= bound1) & (filteredData["HEADER_TIME_STAMP"] <= bound2)].head()
```
